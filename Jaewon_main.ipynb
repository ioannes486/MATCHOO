```python
def recommend_place(prompt):
    model_engine = 'text-davinci-002'
    max_length = 50
    max_tokens = 10
    temperature = 0.5
    prompt = prompt[:max_length] + " tourist attractions in Seoul"
    completions = openai.Completion.create(
        engine=model_engine,
        prompt=prompt,
        max_tokens=max_tokens,
        n=1,
        stop=None,
        temperature=temperature,
    )
    message = completions.choices[0].text
    return message.strip()

prompt = "Seoul tourist attractions"
recommended_place = recommend_place(prompt)
print(f"Recommended place: {recommended_place}")
```
